## What is continent-wise population ratio in 1970?

```python
populatio_1970 = df_population.groupby('Continent').agg({'1970 Population':'sum'})
pop_1970_percentage = populatio_1970.apply(lambda x:  100*x / x.sum())
pop_1970_percentage
```

## What is continent-wise population ratio in 2022?

```python
populatio_2022 = df_population.groupby('Continent').agg({'2022 Population':'sum'})
pop_2022_percentage = populatio_2022.apply(lambda x:  100*x / x.sum())
pop_2022_percentage

plt.pie(pop_2022_percentage)
plt.show


```
