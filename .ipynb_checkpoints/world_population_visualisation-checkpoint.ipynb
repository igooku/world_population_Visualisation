## What is continent-wise population ratio in 1970?

```python
populatio_1970 = df_population.groupby('Continent')['1970 Population'].sum()
pop_1970_percentage = populatio_1970.div(populatio_1970.sum()).mul(100)
pop_1970_percentage

plt.pie(pop_1970_percentage, labels = pop_1970_percentage.index, autopct='%1.1f%%', explode = (0.1, 0.1, 0.1, 0.1, 0.5, 0.1) ,wedgeprops = {"edgecolor": "black"})
plt.title("Population of different Continent in 1970 (Millions)")
plt.show()
populatio_1970
```

## What is continent-wise population ratio in 2022?

```python
population_2022 = df_population.groupby('Continent')['2022 Population'].sum()
pop_2022_percentage = population_2022.div(population_2022.sum()).mul(100)
pop_2022_percentage

plt.pie(pop_2022_percentage, labels = pop_2022_percentage.index, autopct='%1.1f%%', explode = (0.1, 0.1, 0.1, 0.1, 0.5, 0.1) ,wedgeprops = {"edgecolor": "black"})
plt.title("Population of different Continent in 2022(Millions)")
plt.show()
population_2022
```

## Top five most populated countries?

```python
countries_population = df_population.groupby('Country/Territory')['2022 Population'].sum()
df_most_poluted = countries_population.sort_values(ascending=False).head(5)
df_most_poluted = df_most_poluted.div(10**9)

ax = df_most_poluted.plot(kind = 'bar', edgecolor = 'black', rot = 0, figsize = [8,6])
ax.bar_label(ax.containers[0], fontsize = 10)

plt.xlabel('Top five most populated countries(Millions)')
plt.show()
df_most_poluted
```

## Top five least populated countries?

```python
countries_population_l = df_population.groupby('Country/Territory')['2022 Population'].sum()
df_least_poluted = countries_population_l.sort_values(ascending=True).head(5)

ax = df_least_poluted.plot(kind = 'bar', edgecolor = 'black', rot = 0, figsize = [8,6])
ax.bar_label(ax.containers[0], fontsize = 12)

plt.xlabel('Top five Least populated countries')
plt.show()
df_least_poluted
```

## How many country in each continent?

```python
df_countries = df_population['Continent'].value_counts()

ax = df_countries.plot(kind = 'bar', edgecolor = 'black', rot = 0, figsize = [8,6])
ax.bar_label(ax.containers[0], fontsize = 12)

plt.xlabel('Countries in each continent')
plt.show()
```

## Pakistan population over the years.

```python
# find pakistan row in dataset
specific_country = ['Pakistan']
df_population[df_population['Country/Territory'].isin(specific_country)]
```

```python
pop_pak = df_population[df_population['Country/Territory']== 'Pakistan']
#list of columns needed
pop_cols= [x for x in pop_pak.columns if x.endswith('Population')]
#get only relevent columns
pop_data = pop_pak[pop_cols]
#split only year from all populations year

pop_data.columns = [x.split()[0] for x in pop_data.columns]
pop_data = pop_data.T.sort_index()

pop_data.columns = ['Pakistan Population']
pop_data
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = [14,6]);

pop_data.plot(ax = ax[0,0]);
pop_data.plot.bar(rot = 0, ax = ax[0,1]);

pop_data.div(10**6).plot(ax=ax[1,0], title=f'{pop_data.columns[0]}: Millions');
pop_data.div(10**6).plot.bar(rot=0, ax=ax[1,1], title=f'{pop_data.columns [0]}: Millions');

plt.tight_layout();
```
